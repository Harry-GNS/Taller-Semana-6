Ejercicio 1 
Tercer Polinomio de Taylor para cos(x) si x<sub>0</sub> = 0

```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial

# Definir el rango de x
x = np.linspace(-2*np.pi, 2*np.pi, 1000)

# Funci√≥n original
def original_function(x):
    return np.cos(x)

# Funci√≥n para calcular el polinomio de Taylor de orden n para cos(x) en x0=0
def taylor_cos(x, n):
    result = 0
    for i in range(n+1):
        # Solo los t√©rminos pares aparecen en la serie de cos(x)
        if i % 2 == 0:
            # El signo alterna: 1, -1, 1, -1, ...
            sign = (-1)**(i//2)
            result += sign * (x**i) / factorial(i)
    return result

# Crear la gr√°fica
plt.figure(figsize=(12, 8))

# Graficar la funci√≥n original
plt.plot(x, original_function(x), 'k-', label='cos(x)', linewidth=2)

# Graficar los polinomios de Taylor de diferentes √≥rdenes
orders = [2, 4, 6, 10, 12, 14, 15]
colors = ['r', 'g', 'b', 'm', 'c', 'y', 'orange']

for i, order in enumerate(orders):
    taylor_approx = taylor_cos(x, order)
    plt.plot(x, taylor_approx, f'{colors[i % len(colors)]}--',
             label=f'Taylor orden {order} en $x_0=0$')

# Configurar la gr√°fica
plt.axhline(y=0, color='gray', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='gray', linestyle='-', alpha=0.3)
plt.grid(True, alpha=0.3)
plt.legend(loc='upper right')
plt.title('Funci√≥n cos(x) y sus aproximaciones de Taylor en $x_0=0$')
plt.xlabel('x')
plt.ylabel('y')

# Limitar el rango del eje y para mejor visualizaci√≥n
plt.ylim(-2, 2)

# A√±adir marcas para m√∫ltiplos de œÄ en el eje x
ticks = np.arange(-2*np.pi, 2*np.pi + 0.1, np.pi/2)
labels = [r'$-2\pi$', r'$-3\pi/2$', r'$-\pi$', r'$-\pi/2$', r'$0$',
          r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$']
plt.xticks(ticks, labels)

plt.show()

# Mostrar las f√≥rmulas de los primeros polinomios de Taylor
print("Polinomios de Taylor para cos(x) en x‚ÇÄ = 0:")
print("Orden 2: cos(x) ‚âà 1 - x¬≤/2")
print("Orden 4: cos(x) ‚âà 1 - x¬≤/2 + x‚Å¥/24")
print("Orden 6: cos(x) ‚âà 1 - x¬≤/2 + x‚Å¥/24 - x‚Å∂/720")
print("Orden 10: cos(x) ‚âà 1 - x¬≤/2 + x‚Å¥/24 - x‚Å∂/720 + x‚Å∏/40320 - x¬π‚Å∞/3628800")
```
