```python

import numpy as np
import matplotlib.pyplot as plt

# Definir la funcin original
def ln_x(x):
    return np.log(x)

# Definir los polinomios de Taylor de diferentes rdenes
def taylor_ln_1(x):  # Primer orden
    return (x-1)

def taylor_ln_2(x):  # Segundo orden
    return (x-1) - (x-1)**2/2

def taylor_ln_3(x):  # Tercer orden
    return (x-1) - (x-1)**2/2 + (x-1)**3/3

def taylor_ln_4(x):  # Cuarto orden
    return (x-1) - (x-1)**2/2 + (x-1)**3/3 - (x-1)**4/4

def taylor_ln_5(x):  # Quinto orden
    return (x-1) - (x-1)**2/2 + (x-1)**3/3 - (x-1)**4/4 + (x-1)**5/5

# Crear un rango de valores x para graficar
x = np.linspace(0.5, 2, 1000)

# Crear la figura y los ejes
plt.figure(figsize=(10, 6))

# Graficar la funcin original
plt.plot(x, ln_x(x), 'k-', linewidth=2, label='ln(x)')

# Graficar los polinomios de Taylor
plt.plot(x, taylor_ln_1(x), 'r--', label='Orden 1: (x-1)')
plt.plot(x, taylor_ln_2(x), 'g--', label='Orden 2: (x-1) - (x-1)/2')
plt.plot(x, taylor_ln_3(x), 'b--', label='Orden 3: (x-1) - (x-1)/2 + (x-1)/3')
plt.plot(x, taylor_ln_4(x), 'm--', label='Orden 4')
plt.plot(x, taylor_ln_5(x), 'c--', label='Orden 5')

# Aadir una lnea vertical en x = 1
plt.axvline(x=1, color='gray', linestyle='-', alpha=0.3)

# Configurar el grfico
plt.grid(True, alpha=0.3)
plt.title('Funcin ln(x) y sus aproximaciones de Taylor alrededor de x = 1')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.axhline(y=0, color='gray', linestyle='-', alpha=0.3)

# Mostrar el grfico
plt.tight_layout()
plt.show()
```
